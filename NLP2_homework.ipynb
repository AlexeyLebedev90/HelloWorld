```python
# NLP2_1
import re
import io, sys

addresses = []

input_stream = io.TextIOWrapper(sys.stdin.buffer, encoding='utf-8')
output_stream = io.TextIOWrapper(sys.stdout.buffer, encoding='utf-8')

for _ in range(int(input_stream.readline())):
    for address in re.findall(r"[\w\.]+@[\w\.]+\w",input_stream.readline().strip()):
        if address not in addresses:
            addresses.append(address)

print(";".join(sorted(addresses)))
```

```python
# NLP2_2
import re

domain_set = set()
domain_regex = re.compile(r'https?://([a-zA-Z0-9][a-zA-Z0-9-]*\.[a-zA-Z0-9-.]+)[/\?]?')

for line in range(int(input())):
    for m in domain_regex.findall(input()):
        if m.startswith("www.") or m.startswith("ww2."):
            domain_set.add(m[4:])
        else:
            domain_set.add(m)

domains = sorted(domain_set)
print(";".join(domains))
```

```python
# NLP2_3
import pandas as pd
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import PorterStemmer 
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('labeled.csv')
df
```

```python
stem = PorterStemmer()

df['stem_comm'] = df['comment'].apply(lambda x: " ".join([stem.stem(word) for word in x.split()]))
df
```

```python
lem = WordNetLemmatizer()

df['lem_comm'] = df['comment'].apply(lambda x: " ".join([lem.lemmatize(word) for word in x.split()]))
df
```

```python
vec = CountVectorizer()
df1 = vec.fit_transform(df['lem_comm'])
print(df1)
```

```python
# NLP2_4

#Разделение обучения и тестирования
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.comment, df.toxic, test_size = 0.25,random_state = 42)
X_train.reset_index(drop = True,inplace = True)
X_test.reset_index(drop = True,inplace = True)
y_train.reset_index(drop = True,inplace = True)
y_test.reset_index(drop = True,inplace = True)
```

```python
#Установка векторизатора TF-IDF 
from sklearn.feature_extraction.text import TfidfVectorizer
vec_tfidf = TfidfVectorizer()
vec_tfidf_f = vec_tfidf.fit_transform(X_train)
```

```python
#модель классификации XG-boost с векторизатором TF-IDF
from xgboost import XGBClassifier as xgbc
xgbc_f = xgbc(n_estimators=200).fit(vec_tfidf_f,y_train)
```

```python
y_pred_xgb = xgbc_f.predict(vec_tfidf_f)
```

```python
print(y_test, y_pred_xgb)
```
